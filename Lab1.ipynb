```python
import random

def calculate_sum (number, b =[]):
    summa =0
    o = random.sample(range(1, len(b) - 1), int(number*len(b)))
    for i in range(len(o)):
        summa = (summa + b[o[i]])%(amax +1)
    return summa
```

```python
import time


def BruteForce (s, v=[]):
    x=[]
    n = len(v)
    get_bin = lambda x, n: format(x, 'b').zfill(n)
    start = time.time()
    for i in range(pow(2,n)):
        a = [int(x) for x in list(get_bin(i, n))]
        summ = sum(v*a)%(amax+1)
        if summ == s:
            end = time.time()
            x.append((end - start))
        if len(x) == 5:
            break
    return x
```
